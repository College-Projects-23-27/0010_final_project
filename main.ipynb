## Main Notebook

```python
neighborhoods = {}
```

```python
from util.graphical_neighborhood import graph_data
import primary_care_access
import obesity_rate
import lead_level
```

```python
all_healthcare_providers = primary_care_access.providers
for i in range(all_healthcare_providers[0]):
    try:
        neighborhoods[all_healthcare_providers[0][i]].append(i+1)
    except:
        neighborhoods[all_healthcare_providers[0][i]] = [i+1]
graph_data(all_healthcare_providers[0], all_healthcare_providers[1])
```

```python
blood_lead_levels_dataframe = lead_level.lead_levels
for i in range(blood_lead_levels_dataframe[0]):
    try:
        neighborhoods[blood_lead_levels_dataframe[0][i]].append(i+1)
    except:
        neighborhoods[blood_lead_levels_dataframe[0][i]] = [i+1]
graph_data(blood_lead_levels_dataframe[0], blood_lead_levels_dataframe[1])
```

```python
obesity_rates_dataframe = obesity_rate.pittsburgh_obesity_areas
for i in range(obesity_rates_dataframe[0]):
    try:
        neighborhoods[obesity_rates_dataframe[0][i]].append(i+1)
    except:
        neighborhoods[obesity_rates_dataframe[0][i]] = [i+1]
graph_data(obesity_rates_dataframe[0], obesity_rates_dataframe[1])
```

```python
working = {}
for i in neighborhoods:
    # If it shows up on at least 2 lists, giving room for the blood lead level rankings
    if len(neighborhoods[i]) >= 2:
        working[i] = neighborhoods[i].sum() / 3

final_rankings = sorted(working.items())
graph_data(final_rankings, [working[i] for i in final_rankings])
```
